```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data (Replace with your actual data)
descriptors = ['SIFT', 'ORB', 'BRISK', 'FAST']
svm_accuracies = [0.042, 0.024, 0.014, 0.017]
cnn_accuracies = [0.066, 0.020, 0.031, 0.020]

# Set the width of each bar
bar_width = 0.35

# Set the x-axis positions for each group of bars
x = np.arange(len(descriptors))

# Create the figure and axis objects
fig, ax = plt.subplots(figsize=(10, 6))
# Create the bars for SVM accuracies
rects1 = ax.bar(x - bar_width/2, svm_accuracies, bar_width, label='SVM')

# Create the bars for CNN accuracies
rects2 = ax.bar(x + bar_width/2, cnn_accuracies, bar_width, label='CNN')

# Add labels and title
ax.set_xlabel('Image Descriptors', fontsize=18)
ax.set_ylabel('Accuracy', fontsize=18)
ax.set_title('Accuracy of Image Descriptors with SVM and CNN', fontsize=20)
ax.set_xticks(x)
ax.set_xticklabels(descriptors, fontsize=16)
ax.tick_params(axis='y', labelsize=18)

# Add legend
ax.legend(fontsize=20)

# Add value labels on top of each bar
# for i, rect1, rect2 in zip(x, rects1, rects2):
#     height1 = rect1.get_height()
#     height2 = rect2.get_height()
    # ax.annotate(f'{height1:.2f}',
    #             xy=(rect1.get_x() + rect1.get_width() / 2, height1),
    #             xytext=(0, 3),
    #             textcoords='offset points',
    #             ha='center', va='bottom')
    # ax.annotate(f'{height2:.2f}',
    #             xy=(rect2.get_x() + rect2.get_width() / 2, height2),
    #             xytext=(0, 3),
    #             textcoords='offset points',
    #             ha='center', va='bottom')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the chart
plt.show()
```
