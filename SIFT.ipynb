```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv.imread('./test_data/ILSVRC2012_val_00000001_n01751748.JPEG')
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
sift = cv.SIFT_create(64)
keypoints = sift.detect(gray, None)

kp_img = cv.drawKeypoints(gray, keypoints, img)

plt.imshow(img)
```

```python
kp1, des1 = sift.compute(gray, keypoints)
```

```python
print(np.array(kp1).shape)
print(des1.shape)
```
